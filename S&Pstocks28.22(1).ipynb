```python
import math
import numpy as np
import pandas as pd
from datetime import datetime
import time
import pandas_datareader as web
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from keras.models import Sequential
from keras.layers import Dense, LSTM
#import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# Selecting columns 'Date' and 'Close' from the pandas DataFrame 'pandas_df'.
selected_df = pandas_df[['Date', 'Close']]

# Scale only the 'Close' column
scaler = MinMaxScaler(feature_range=(0, 1))
selected_df['Close'] = scaler.fit_transform(selected_df[['Close']])
```

```python
# Displaying the first few rows of the selected DataFrame 'selected_df' after data selection and scaling.
selected_df.head()
```

```python
# Filtering the 'Close' column from the DataFrame 'selected_df' and assigning it to a new DataFrame 'data'.
data =selected_df.filter(['Close'])
```

```python
# Displaying the DataFrame 'data' containing the filtered 'Close' column.
data
```

```python
# Converting the DataFrame 'data' into a NumPy array and assigning it to the variable 'dataset'.
dataset = data.values
```

```python
# Calculating the length of the training data by taking 80% of the total dataset length and rounding up to the nearest integer.
training_data_len = math.ceil( len(dataset) * .8 )
```

```python
# Displaying the length of the training data.
training_data_len
```

```python
# Creating a MinMaxScaler object with a feature range between 0 and 1.
scaler = MinMaxScaler(feature_range=(0,1))
# Scaling the dataset using the MinMaxScaler.
dataset = scaler.fit_transform(dataset)
```

```python
#scaled_data
dataset
```

```python
# Slicing the scaled dataset to obtain the training data, which consists of the first 'training_data_len' rows and all columns.
train_data = dataset[0:training_data_len , :]
```

```python
# Initializing empty lists to store the input features (x_train) and target values (y_train).
x_train =[]
y_train =[]
```

```python
# Generating training data by iterating through the train_data array.
for i in range(30, len(train_data)):
     # Appending the previous 30 data points as input features (x_train).
    x_train.append(train_data[i-30:i,0])
    # Appending the next data point as the target value (y_train).
    y_train.append(train_data[i,0])
```

```python
# Converting the lists of input features (x_train) and target values (y_train) into NumPy arrays.
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Reshaping the input features (x_train) to be compatible with LSTM input shape: (samples, time steps, features).
x_train =np.reshape(x_train,(x_train.shape[0], x_train.shape[1],1))
# Displaying the shape of the reshaped x_train.
x_train.shape
```

```python
# Creating a Sequential model.
model = Sequential()
# Adding the first LSTM layer with 50 units and returning sequences.
model.add(LSTM(50, return_sequences = True, input_shape = (x_train.shape[1], 1)))
# Adding the second LSTM layer with 50 units and not returning sequences.
model.add(LSTM(50, return_sequences = False))
# Adding a Dense layer with 5 units.
model.add(Dense(5))
# Adding a Dense layer with 1 unit (output layer).
model.add(Dense(1))
```

```python
#model.compile(optimizer='adam', loss = 'mean_squared_error')
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fitting the model using the training data.
# Training the model with a batch size of 5 and for 30 epochs.
model.fit(x_train, y_train, batch_size = 5, epochs = 30)
```

```python
# Selecting the test data, which includes the last 30 data points from the dataset.
test_data = dataset[training_data_len -30: , :]
```

```python
# Initializing an empty list to store the input features for the test data.
x_test = []
# Assigning the y_test variable to the test data without the input features.
y_test = dataset[training_data_len:, :]
```

```python
# Generating input features for the test data by iterating through the test_data array.
for i in range(30,len(test_data)):
      # Appending the previous 30 data points as input features for each test data point.
    x_test.append(test_data[i-30:i, 0])
```

```python
# Converting the list of input features for the test data (x_test) into a NumPy array.
x_test = np.array(x_test)
```

```python
# Reshaping the input features for the test data (x_test) to be compatible with LSTM input shape: (samples, time steps, features).
x_test = np.reshape(x_test,(x_test.shape[0], x_test.shape[1],1))
```

```python
# Generating predictions using the trained model on the test data.
predictions = model.predict(x_test)
```

```python
# Inverse scaling the predictions to obtain the actual predicted values.
predictions = scaler.inverse_transform(predictions)
```

```python
# Calculating the Root Mean Squared Error (RMSE) between the actual target values and the predictions.
rmse = np.sqrt(mean_squared_error(scaler.inverse_transform(y_test),predictions))
# Calculating the R-squared (R2) value between the actual target values and the predictions.
r2_value = r2_score(scaler.inverse_transform(y_test),predictions)
```

```python
# Printing the Root Mean Squared Error (RMSE) and R-squared (R2) value.

print("Root Mean Square Error: ", rmse)
print("R^2 Value: ", r2_value)
```

```python
# Separating the training and validation data.
train = data[:training_data_len]
valid = data[training_data_len:]
# Adding a new column 'Predictions' to the validation data DataFrame and assigning the predicted values.
valid['Predictions'] = predictions
```

```python
# Plotting the actual 'Close' prices for the training data, along with the actual and predicted 'Close' prices for the validation data.
plt.figure(figsize=(16,8))
plt.title('LSTM Deep Learning Neural Networks Model')
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close Price USD($)', fontsize=18)
plt.plot(train['Close'])# Plotting the training data
plt.plot(valid[['Close', 'Predictions']]) # Plotting the actual and predicted values for the validation data
plt.legend(['Train', 'Valid', 'Predictions'], loc = 'lower right')
plt.show()
```
